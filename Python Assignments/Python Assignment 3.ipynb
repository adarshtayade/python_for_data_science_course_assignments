#### 11. Count the average time of work (hours-per-week) for those who earn a little and a lot (salary) for each country (native-country). What will these be for Japan? 

```python
countries = df['native-country'].unique()
# For those who earn less
for cou in countries:
    mean_hour = df[df['Salary_range'] == '<=50K']['hours-per-week'].mean()
    print(cou , " - " , mean_hour)
```

```python
# For those who earn a lot
for cou in countries:
    mean_hour = df[df['Salary_range'] == '>50K']['hours-per-week'].mean()
    print(cou , " - " , mean_hour)
```
