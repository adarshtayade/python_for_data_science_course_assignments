### 2. Grouping and Aggregations 

Dataset - https://raw.githubusercontent.com/datagy/data/main/sales.csv 

#### Load data and group by Region 

```python
data = pd.read_csv(r'https://raw.githubusercontent.com/datagy/data/main/sales.csv')
data
```

```python
data.groupby(['region'])
```

#### Counting number of groups

```python
len(data.groupby(['region']))
```

#### Accessing the Groups in a GroupBy object

```python
(data.groupby(['region'])).groups
```

#### Accessing only Group Names of a GroupBy Object 

```python
(data.groupby('region')).groups.keys()
```

#### Selecting a Group of south region 

```python
data[data['region'] == 'South']
```

#### Findout average sales per region

```python
avg_sale = data.groupby(['region'])['sales'].mean()
avg_sale
```

#### Applying Multiple Aggregations Using Pandas GroupBy - Find out Mean, median and variance of sales per region

```python
data.groupby('region')['sales'].aggregate(['mean', 'median', 'var'])
```

#### Calculate percentage of region's sales per row with respect to total sale of the region

```python
data['sales_per'] = (data['sales']/data['sales'].sum())*100
data
```

#### Filtering Rows Where the Group's Average Sale Price is Less Than 20,000

```python
fil = avg_sale[avg_sale < 20000]
filtered = data[data['region'].isin(fil.index)]
filtered
```

#### 10. Calculate the sum of all sales broken out by 'region' and by 'gender'

```python
data.groupby(['region', 'gender'])['sales'].sum()
```

#### 11. Ranking Sales by Region and by Gender

```python
data['rank'] = data.groupby(['region', 'gender'])['sales'].rank(ascending = False)
data
```
