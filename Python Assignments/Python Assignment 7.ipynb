```python
# Importing libraries
import numpy as np
import pandas as pd
```

### Load NYCFlights dataset

```python
data = pd.read_csv('flight_data.csv')
data.head(5)
```

#### 1. List all flights scheduled for 1 Jan

```python
data[data['time_hour'].str.startswith('1/1')]
```

#### 2. Find flights that were not delayed more than 2 hours

```python
data[data['dep_delay'] < 2]
```

#### 3. Find all flights

##### a. Had an delay of 2 or more hours

```python
data[data['dep_delay'] >= 2]
```

##### b. Flew to Houston (IAH or HOU)

```python
data[data['dest'] == ('IAH' or 'HOU')]
```

##### c. Were operated by Unites, American or Delta

```python
data['carrier'].unique()
```

```python
data[data['carrier'] == ('UA' or 'AA' or 'DL')]
```

##### d. Departed in summer (March, April , May)

```python
data[data['month'] == (3 or 4 or 5)]
```

##### e. Arrived more than 2 hours late but didn’t leave late

```python
data[(data['arr_delay'] > 2) & (data['dep_delay'] <= 0)]
```

##### f. Departed between midnight and 6.00 am

```python
data[(data['dep_time'] > 000) & (data['dep_time'] <= 600)]
```

#### 4. Sort flights to find most delayed flights

```python
data['dep_delay'].rank(ascending = False)
```
